```python
import urllib.request 

wiki = "https://www.flipkart.com/fogg-fresh-oriental-body-spray-men/product-reviews/itmfy4qdjhtxxzkh?page=1&pid=DEODPAB82VDQ2HHA"

page = urllib.request.urlopen(wiki) 
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page, "lxml")

print (soup.prettify())
```

```python
elements = soup.find_all("span", class_="_2jRR3v".split())
print("\n".join("{} {}".format(el['class'], el.get_text()) for el in elements))
```
