```python
import urllib.request 

wiki = "https://www.goodreads.com/book/show/10872085-only-time-will-tell?ac=1&from_search=true"

page = urllib.request.urlopen(wiki) 
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page, "lxml")

print (soup.prettify())
```

```python
elements = soup.find_all("span", class_ ="reviewTextContainer1937805413".split())
print("\n".join("{} {}".format(el['class'], el.get_text()) for el in elements))
```
